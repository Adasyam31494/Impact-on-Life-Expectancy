```python
import pandas as pd

# Load the dataset
file_path = "Life_Expectancy_Data.csv"
df = pd.read_csv(file_path)

# Clean column names by stripping leading/trailing whitespace
df.columns = df.columns.str.strip()

# 1. Population Size – Categorize population into Small, Medium, Large
def categorize_population(pop):
    if pd.isna(pop):
        return None
    elif pop < 30000:
        return "Small"
    elif 30000 <= pop < 100000:
        return "Medium"
    else:
        return "Large"

df["Population Size"] = df["Population"].apply(categorize_population)

# 2. Lifestyle – Average of Alcohol consumption and BMI
df["Lifestyle"] = df[["Alcohol", "BMI"]].mean(axis=1)

# 3. Economy – Product of Population and GDP
df["Economy"] = df["Population"] * df["GDP"]

# 4. Death Ratio – Adult Mortality / Infant deaths
df["Death Ratio"] = df.apply(
    lambda row: row["Adult Mortality"] / row["infant deaths"] if row["infant deaths"] != 0 else None,
    axis=1
)

# Optional: Save to a new CSV
# df.to_csv("Enhanced_Life_Expectancy_Data.csv", index=False)

# Display the first few rows
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Drop rows where target variable 'Life expectancy' is missing
df_clean = df.dropna(subset=["Life expectancy"])

# Check for remaining missing values
missing_counts = df_clean.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]

# Impute missing values:
# - Numerical columns: fill with median
for col in missing_counts.index:
    if df_clean[col].dtype in [np.float64, np.int64]:
        df_clean[col].fillna(df_clean[col].median(), inplace=True)
    else:
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

# Identify numerical predictors
numerical_predictors = df_clean.select_dtypes(include=[np.number]).drop(columns=["Life expectancy"]).columns

# Generate scatter plots and apply log transformation if necessary
log_transformed = []
for col in numerical_predictors:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df_clean[col], y=df_clean["Life expectancy"])
    plt.title(f'Scatter plot: {col} vs Life Expectancy')
    plt.xlabel(col)
    plt.ylabel("Life Expectancy")
    plt.tight_layout()
    plt.show()

    # Check if log transformation helps (right-skewed data)
    if (df_clean[col] > 0).all() and df_clean[col].skew() > 1:
        df_clean[f"log_{col}"] = np.log(df_clean[col])
        log_transformed.append(col)

log_transformed
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Life_Expectancy_Data.csv")
df.columns = df.columns.str.strip()  # Clean column names

# Drop rows where Life Expectancy is missing
df_clean = df.dropna(subset=["Life expectancy"])

# Fill missing values
for col in df_clean.columns:
    if df_clean[col].dtype in [np.float64, np.int64]:
        df_clean[col].fillna(df_clean[col].median(), inplace=True)
    else:
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

# Recreate important engineered features
df_clean["Economy"] = df_clean["Population"] * df_clean["GDP"]
df_clean["Death Ratio"] = df_clean.apply(
    lambda row: row["Adult Mortality"] / row["infant deaths"] if row["infant deaths"] != 0 else np.nan,
    axis=1
)

# Apply log transformation to skewed columns
columns_to_log_transform = [
    'Adult Mortality', 'HIV/AIDS', 'GDP', 'Population',
    'thinness  1-19 years', 'thinness 5-9 years', 'Economy', 'Death Ratio'
]

for col in columns_to_log_transform:
    if (df_clean[col] > 0).all():
        df_clean[f"log_{col}"] = np.log(df_clean[col])

# Drop original columns after log-transforming
df_task2_final = df_clean.drop(columns=columns_to_log_transform)

# ------------------------------------
# Begin Task 5: Data Analysis
# ------------------------------------

# Create a working copy for analysis
df_analysis = df_task2_final.copy()

# Restore 'Population' and 'GDP' for density analysis
df_analysis["Population"] = df_clean["Population"]
df_analysis["GDP"] = df_clean["GDP"]

# Q1: Should countries with life expectancy < 65 increase healthcare expenditure?
low_life_exp = df_analysis[df_analysis["Life expectancy"] < 65]
high_life_exp = df_analysis[df_analysis["Life expectancy"] >= 65]

exp_low = low_life_exp["Total expenditure"]
exp_high = high_life_exp["Total expenditure"]

# T-test between expenditure groups
t_stat, p_val_exp = stats.ttest_ind(exp_high, exp_low, equal_var=False)

# Q2: Schooling vs Life Expectancy
school_corr, p_val_school = stats.pearsonr(df_analysis["Schooling"], df_analysis["Life expectancy"])

# Q3: Alcohol vs Life Expectancy
alcohol_corr, p_val_alcohol = stats.pearsonr(df_analysis["Alcohol"], df_analysis["Life expectancy"])

# Q4: Population Density (proxy) vs Life Expectancy
df_analysis["Density Proxy"] = df_analysis["Population"] / df_analysis["GDP"]
density_corr, p_val_density = stats.pearsonr(df_analysis["Density Proxy"], df_analysis["Life expectancy"])

# Final Output Summary
results = {
    "Q1_Healthcare_Expenditure": {
        "Mean Expenditure (Life Exp <65)": round(exp_low.mean(), 2),
        "Mean Expenditure (Life Exp >=65)": round(exp_high.mean(), 2),
        "T-Test p-value": round(p_val_exp, 5)
    },
    "Q2_Schooling_Correlation": {
        "Correlation": round(school_corr, 2),
        "p-value": round(p_val_school, 5)
    },
    "Q3_Alcohol_Correlation": {
        "Correlation": round(alcohol_corr, 2),
        "p-value": round(p_val_alcohol, 5)
    },
    "Q4_Density_Correlation": {
        "Correlation": round(density_corr, 2),
        "p-value": round(p_val_density, 5)
    }
}

# Print results
for q, metrics in results.items():
    print(f"\n{q.replace('_', ' ')}:")
    for key, value in metrics.items():
        print(f"  {key}: {value}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

# Load dataset
df = pd.read_csv("Life_Expectancy_Data.csv")
df.columns = df.columns.str.strip()

# Clean and prepare dataset
df_clean = df.dropna(subset=["Life expectancy"])

# Impute missing values
for col in df_clean.columns:
    if df_clean[col].dtype in [np.float64, np.int64]:
        df_clean[col].fillna(df_clean[col].median(), inplace=True)
    else:
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

# Feature engineering
df_clean["Economy"] = df_clean["Population"] * df_clean["GDP"]
df_clean["Death Ratio"] = df_clean.apply(
    lambda row: row["Adult Mortality"] / row["infant deaths"] if row["infant deaths"] != 0 else np.nan,
    axis=1
)

# Apply log transformation
columns_to_log_transform = [
    'Adult Mortality', 'HIV/AIDS', 'GDP', 'Population',
    'thinness  1-19 years', 'thinness 5-9 years', 'Economy', 'Death Ratio'
]

for col in columns_to_log_transform:
    if (df_clean[col] > 0).all():
        df_clean[f"log_{col}"] = np.log(df_clean[col])

# Final dataset
df_model = df_clean.drop(columns=columns_to_log_transform)
X = df_model.select_dtypes(include=[np.number]).drop(columns=["Life expectancy"])
y = df_model["Life expectancy"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
y_pred_all = model.predict(X)

# Evaluation metrics
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def evaluate(y_true, y_pred):
    return {
        "RMSE": rmse(y_true, y_pred),
        "MAPE": mean_absolute_percentage_error(y_true, y_pred),
        "R2 Score": r2_score(y_true, y_pred)
    }

performance = {
    "Train": evaluate(y_train, y_pred_train),
    "Test": evaluate(y_test, y_pred_test),
    "Overall": evaluate(y, y_pred_all)
}

# Residual Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred_all, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', label='Ideal Unity Line')
plt.plot([y.min(), y.max()], [y.min() + 5, y.max() + 5], 'y--', label='+5 Years')
plt.plot([y.min(), y.max()], [y.min() - 5, y.max() - 5], 'y--')
plt.plot([y.min(), y.max()], [y.min() + 10, y.max() + 10], 'r--', label='+10 Years')
plt.plot([y.min(), y.max()], [y.min() - 10, y.max() - 10], 'r--')
plt.xlabel("Actual Life Expectancy")
plt.ylabel("Predicted Life Expectancy")
plt.title("Residual Scatter Plot")
plt.legend()
plt.tight_layout()
plt.show()

# Residual Histogram
residuals = y - y_pred_all
plt.figure(figsize=(7, 4))
plt.hist(residuals, bins=30, edgecolor='black')
plt.title("Residual Histogram")
plt.xlabel("Residual (Actual - Predicted)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X, y, cv=cv, scoring='r2')

# Box plot of CV scores
plt.figure(figsize=(6, 4))
plt.boxplot(cv_scores, vert=False)
plt.title("Cross-Validation R² Score (5-Fold)")
plt.xlabel("R² Score")
plt.tight_layout()
plt.show()

# Print performance summary
print("\nModel Performance Summary:")
for section, metrics in performance.items():
    print(f"\n{section} Set:")
    for key, value in metrics.items():
        print(f"  {key}: {value:.4f}")

print(f"\nCross-Validation R² Mean: {np.mean(cv_scores):.4f}")
print(f"Cross-Validation R² Std Dev: {np.std(cv_scores):.4f}")
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Define model
model = LinearRegression()

# Recursive feature elimination
selector = RFE(model, n_features_to_select=1, step=1)
selector = selector.fit(X, y)

# Rank features (1 = best)
feature_ranking = pd.DataFrame({
    'Feature': X.columns,
    'Rank': selector.ranking_
}).sort_values(by='Rank')
print(feature_ranking)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

def evaluate_feature_subset(features_subset):
    X_sel = X[features_subset]
    model = LinearRegression()
    model.fit(X_sel, y)
    y_pred = model.predict(X_sel)

    residuals = y - y_pred
    within_5 = np.all(np.abs(residuals) <= 5)
    within_10 = np.all(np.abs(residuals) <= 10)

    return {
        "features": features_subset,
        "count": len(features_subset),
        "RMSE": np.sqrt(mean_squared_error(y, y_pred)),
        "within_5": within_5,
        "within_10": within_10
    }

results = []
for i in range(1, len(feature_ranking) + 1):
    selected = feature_ranking['Feature'].iloc[:i].tolist()
    res = evaluate_feature_subset(selected)
    results.append(res)
    if res["within_10"]:
        break  # Stop if all predictions are within ±10 years

# Display result
for r in results:
    print(f"{r['count']} features: {r['features']}, RMSE={r['RMSE']:.2f}, within ±5: {r['within_5']}, within ±10: {r['within_10']}")
```
